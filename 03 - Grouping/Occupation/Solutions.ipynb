### Step 1. Import the necessary libraries

```python
import numpy as np
import pandas as pd

pd.set_option("display.float_format", '{:.2f}'.format)
pd.set_option("display.max_columns", None)
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

```python
path = '../../data/u.user.csv'
df = pd.read_csv(path, sep="|")
df.head()
```

### Step 3. Assign it to a variable called users.

```python
users = df.copy()
users
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby(by=["occupation"])['age'].mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
users.groupby(by=["occupation","gender"])[["gender"]].count()

rep = {"gender":{"F":0, "M":1}}

ratio = (users.replace(rep).groupby(by=['occupation'])[['gender']].agg(['sum', 'count']).loc[:,('gender','sum')]/
users.replace(rep).groupby(by=['occupation'])[['gender']].agg(['sum', 'count']).loc[:,('gender','count')])

ratio.sort_values(ascending=False)*100
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby(by=['occupation'])['age'].agg(['min', 'max'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(by=['occupation', 'gender'])['age'].mean()
```

### Step 8.  For each occupation present the percentage of women and men

```python
users.groupby(by=['occupation'])['gender'].value_counts(normalize=True)*100
```
