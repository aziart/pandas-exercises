### Step 1. Import the necessary libraries

```python
import numpy as np
import pandas as pd
import datetime
import warnings
warnings.filterwarnings("ignore")

pd.set_option("display.float_format", "{:.2f}".format)
pd.set_option("display.max_columns", None)
```

### Step 2. Import the dataset from this [address](https://github.com/guipsamora/pandas_exercises/blob/master/06_Stats/Wind_Stats/wind.data)

```python
path = "../../data/wind/"
fname = "wind.data"
fdesc = "wind.desc"

with open(path+fdesc) as f:
    a = f.read()
    print(a)

df = pd.read_fwf(path+fname)
df
```

### Step 3. Assign it to a variable called data and replace the first 3 columns by a proper datetime index.

```python

# df.assign(new_col = df['Yr'] + df['Yr'])
# pd.to_datetime(dict(year=df.Yr, month=df.Mo, day=df.Dy))
# pd.to_datetime((df.Yr*10000+df.Mo*100+df.Dy).apply(str),format='%Y%m%d')
display(df.head())

def to_year(a: pd.Series):
    return '19' + a.apply(str)

dt_col = pd.to_datetime(dict(year=to_year(df.loc[:, 'Yr'])
                           , month=df.loc[:, "Mo"]
                           , day=df.loc[:, "Dy"]))
df.assign(date = dt_col)
# or
df.insert(loc=0 #len(df.columns)
          , column='date'
          , value=dt_col
          , allow_duplicates=True)
df
```

### Step 9. Create a DataFrame called loc_stats and calculate the min, max and mean windspeeds and standard deviations of the windspeeds at each location over all the days 

#### A different set of numbers for each location.

### Step 10. Create a DataFrame called day_stats and calculate the min, max and mean windspeed and standard deviations of the windspeeds across all the locations at each day.

#### A different set of numbers for each day.

```python
# df.stack().groupby(df.stack().index).mean()
# df.stack().unstack(level=0).T
tmp = df.melt(ignore_index=False)
tmp.select_dtypes(include=np.number).groupby(tmp.index).agg(['min', 'max', 'mean', 'std'])
```

```python
# create the dataframe
day_stats = pd.DataFrame()

# this time we determine axis equals to one so it gets each row.
day_stats['min'] = df.min(axis = 1) # min
day_stats['max'] = df.max(axis = 1) # max 
day_stats['mean'] = df.mean(axis = 1) # mean
day_stats['std'] = df.std(axis = 1) # standard deviations

print(day_stats.shape)
day_stats.head()
```

### Step 11. Find the average windspeed in January for each location.  
#### Treat January 1961 and January 1962 both as January.

```python
# df_jans = df.melt(ignore_index=False).loc[['1961-01-01', '1962-01-01'],:]
df_jans = df[df.index.month==1].melt(ignore_index=False)
df_jans.groupby(by=['variable'])['value'].mean().sort_values(ascending=False)
```
