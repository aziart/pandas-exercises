The following exercise utilizes data from [UC Irvine Machine Learning Repository](https://archive.ics.uci.edu/ml/datasets/Auto+MPG)

### Step 1. Import the necessary libraries

```python
import numpy as np
import pandas as pd

pd.set_option("display.float_format", "{:.2f}".format)
pd.set_option("display.max_columns", None)
```

### Step 2. Import the first dataset [cars1](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars1.csv) and [cars2](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars2.csv).  

```python
path = "../../data/auto_mpg/"
file_name_1 = "cars1.csv"
file_name_2 = "cars2.csv"
display(pd.read_csv(path+file_name_1))
display(pd.read_csv(path+file_name_2))
```

   ### Step 3. Assign each to a to a variable called cars1 and cars2

```python
cars1 = pd.read_csv(path+file_name_1)
cars2 = pd.read_csv(path+file_name_2)
display(cars1)
display(cars2)
```

### Step 4. Oops, it seems our first dataset has some unnamed blank columns, fix cars1

```python
cars1.dropna(axis=1
             , inplace=True)
cars1
```

### Step 5. What is the number of observations in each dataset?

```python
print(cars1.shape)
print(cars2.shape)
```

### Step 6. Join cars1 and cars2 into a single DataFrame called cars

```python
cars = pd.concat([cars1, cars2])
cars
```

### Step 7. Oops, there is a column missing, called owners. Create a random number Series from 15,000 to 73,000.

```python
owners = np.random.randint(15000, 73000, cars.shape[0])
owners
```

### Step 8. Add the column owners to cars

```python
try:
    cars.insert(loc=len(cars.columns)
                , column="owners"
                , value=owners
                )
except ValueError as e:
    print("exception: {}".format(e))
cars
```
