**3. Pivot Table with Multiple Indexes from Titanic**

**Write a Pandas program to create a Pivot table with multiple indexes from the data set of titanic.csv.**

</div>

```python
import pandas as pd

class Paths:
    titanic_data = '../data/titanic.csv'

paths = Paths()
path = paths.titanic_data
df = pd.read_csv(path)
display(df.head(20))
```

<div class="alert alert-success">

**Solution 03:**
</div>

```python
result = pd.pivot_table(
    data=df
    , index=['pclass', 'sex']
    , values='sibsp'
    , aggfunc='mean'
    , fill_value=0
    , margins=True
    , margins_name='Overall Average'
)

display(result.style.format("{:.2f}").set_caption("Average Number of Siblings/Spouses Aboard by Class"))
display(result.reset_index())
```

**4. Pivot Table: Survival Rate by Gender on Various Classes**

**Write a Pandas program to create a Pivot table and find survival rate by gender on various classes.**

</div>

```python
import pandas as pd

class Paths:
    titanic_data = '../data/titanic.csv'

paths = Paths()
path = paths.titanic_data
df = pd.read_csv(path)
display(df.head(20))
```

<div class="alert alert-success">

**Solution 04:**
</div>

```python
result = pd.pivot_table(
    data=df
    , index=['sex']
    , columns=['class']
    , values='survived'
    , aggfunc='mean'
    , fill_value=0
    , margins=True
    , margins_name='Overall Average'
)

display(result.style.format("{:.2f}").set_caption("Survival Rates by Sex and Class"))
df.pivot_table('survived', index='sex', columns='class')
```
