**3. Pivot Table with Multiple Indexes from Titanic**

**Write a Pandas program to create a Pivot table with multiple indexes from the data set of titanic.csv.**

**Solution 03:**
</div>

```python
result = pd.pivot_table(
    data=df
    , index=['pclass', 'sex']
    , values='sibsp'
    , aggfunc='mean'
    , fill_value=0
    , margins=True
    , margins_name='Overall Average'
)

display(result.style.format("{:.2f}").set_caption("Average Number of Siblings/Spouses Aboard by Class"))
display(result.reset_index())
```

**7. Partition Passengers into Age Categories**

**Write a Pandas program to partition each of the passengers into four categories based on their age.**

**Note: Age categories (0, 10), (10, 30), (30, 60), (60, 80)**

</div>

```python
import pandas as pd

class Paths:
    titanic_data = '../data/titanic.csv'

paths = Paths()
path = paths.titanic_data
df = pd.read_csv(path)
display(df.head(20))
```

<div class="alert alert-success">

**Solution 07:**
</div>

```python
# Write a Pandas program to partition each of the passengers into four categories based on their age.
# Note: Age categories (0, 10), (10, 30), (30, 60), (60, 80)

age_bins = [0, 10, 30, 60, 80]
age_labels = ['0-10', '10-30', '30-60', '60-80']
df['age_category'] = pd.cut(df['age'], bins=age_bins, labels=age_labels, right=False)
result = pd.pivot_table(
    data=df
    , index=['age_category']
    , values='survived'
    , columns='pclass'
    , observed=True
    , aggfunc='mean'
    , fill_value=0
    , margins=True
    , margins_name='Overall Average'
)

result.style.format("{:.2f}").set_caption("Survival Rates by Class and Age Category")
```

**10. Pivot Table: Survival Rate by Gender, Age, and Fare Partition**

**Write a Pandas program to create a Pivot table and find survival rate by gender, age of the different categories of various classes. Add the fare as a dimension of columns and partition fare column into 2 categories based on the values present in fare columns.**

</div>

```python
import pandas as pd

class Paths:
    titanic_data = '../data/titanic.csv'

paths = Paths()
path = paths.titanic_data
df = pd.read_csv(path)
display(df.head(20))
```

<div class="alert alert-success">

**Solution 10:**
</div>

```python
# 10. Pivot Table: Survival Rate by Gender, Age, and Fare Partition
# Write a Pandas program to create a Pivot table and find
# survival rate by gender, age of the different categories of various classes.
# Add the fare as a dimension of columns and partition fare column into 2 categories based on the values present in fare columns.
age_bins = [0, 10, 30, 60, 80]
age_labels = ['0-10', '10-30', '30-60', '60-80']
df['age_category'] = pd.cut(df['age'], bins=age_bins, labels=age_labels, right=False)

fare_bins = [0, 50, 100, 150, 200]
fare_labels = ['0-50', '50-100', '100-150', '150-200']
df['fare_category'] = pd.cut(df['fare'], bins=fare_bins, labels=fare_labels, right=False)
result = pd.pivot_table(
    data=df
    , index=['sex', 'age_category']
    , values='survived'
    , columns='fare_category'
    , aggfunc='mean'
    , fill_value=0
    , margins=True
    , margins_name='Overall Average'
    , observed=True
)

result.style.format("{:.2f}").set_caption("Survival Rates by Class, Age Category, and Gender")
```
