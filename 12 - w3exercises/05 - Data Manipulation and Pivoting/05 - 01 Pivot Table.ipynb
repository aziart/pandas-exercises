**1. Pivot Table with Multiple Indexes**

**Write a Pandas program to create a Pivot table with multiple indexes from a given excel sheet (Salesdata.xlsx).**

</div>

```python
pd.pivot_table?
```

**9. Filtered Pivot Table for Manager "Douglas"**

**Write a Pandas program to create a Pivot table and find the total sale amount region wise, manager wise, sales man wise where Manager = "Douglas".**

</div>

```python
class Paths:
    salesdata = '../data/SaleData.xlsx'

import pandas as pd
import numpy as np
paths = Paths()
path = paths.salesdata
df = pd.read_excel(path)
display(df)
```

<div class="alert alert-success">

**Solution 09:**
</div>

```python
table = pd.pivot_table(
    data=df
    , index=["Region", "Manager", "SalesMan"]
    , values="Sale_amt"
    , aggfunc='sum'
)

display(table.query('Manager == ["Douglas"]'))
# rewrite display(table.query('Manager == ["Douglas"]')) using loc
display(table.loc[table.index.get_level_values('Manager') == 'Douglas'].style.set_caption("Filtered Pivot Table for Manager 'Douglas'"))
```

**10. Region-Wise Television and Home Theater Sold**

**Write a Pandas program to create a Pivot table and find the region wise Television and Home Theater sold.**

```python
class Paths:
    salesdata = '../data/SaleData.xlsx'

import pandas as pd
import numpy as np
paths = Paths()
path = paths.salesdata
df = pd.read_excel(path)
display(df)
```

**11. Maximum Sale Value of Items**

**Write a Pandas program to create a Pivot table and find the maximum sale value of the items.**

</div>

```python
class Paths:
    salesdata = '../data/SaleData.xlsx'

import pandas as pd
import numpy as np
paths = Paths()
path = paths.salesdata
df = pd.read_excel(path)
display(df)
```

<div class="alert alert-success">

**Solution 11:**
</div>

```python
table = pd.pivot_table(
    data=df
    , index=["Item"]
    , values="Sale_amt"
    , aggfunc='max'
    , fill_value=0
    , margins=False
    , margins_name='Total'
    , dropna=True
)
display(table.style.set_caption("Maximum Sale Value of Items").format({'Sale_amt': '{:,.2f}'}))
```
