**Write a Pandas program to create a Pivot table with multiple indexes from a given excel sheet (Salesdata.xlsx).**

```python
class Paths:
    salesdata = '../data/SaleData.xlsx'
```

```python
import pandas as pd
paths = Paths()
path = paths.salesdata
df = pd.read_excel(path)
print(df)
```

<div class="alert alert-success">

**Solution 01:**
</div>

```python
pd.pivot_table?
```

```python
pd.pivot_table(
    data=df
    , index=["Region", "SalesMan"]
    , values=["Units", "Unit_price"]
    , aggfunc={"Units": "sum", "Unit_price": "mean"}
    , fill_value=0
    , margins=True
)
```
