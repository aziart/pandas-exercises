**1. Pivot Table with Multiple Indexes**

**Write a Pandas program to create a Pivot table with multiple indexes from a given excel sheet (Salesdata.xlsx).**

</div>

```python
class Paths:
    salesdata = '../data/SaleData.xlsx'
```

```python
import pandas as pd
paths = Paths()
path = paths.salesdata
df = pd.read_excel(path)
display(df)
```

<div class="alert alert-success">

**Solution 01:**
</div>

```python
pd.pivot_table?
```

```python
pd.pivot_table(
    data=df
    , index=["Region", "SalesMan"]
    , values=["Units", "Unit_price"]
    , aggfunc={"Units": "sum", "Unit_price": "mean"}
    , fill_value=0
    , margins=True
)
```

**2. Region-Wise and Manager-Wise Total Sale Amount**

**Write a Pandas program to create a Pivot table and find the total sale amount region wise, manager wise.**

</div>

```python
class Paths:
    salesdata = '../data/SaleData.xlsx'

import pandas as pd
import numpy as np
paths = Paths()
path = paths.salesdata
df = pd.read_excel(path)
display(df)
```

<div class="alert alert-success">

**Solution 02:**
</div>

```python
display(pd.pivot_table(
    df,index = ["Region","Manager"]
    , values = ["Sale_amt"]
    ,aggfunc='sum'
))
```

**3. Region, Manager, and Salesman-Wise Total Sale Amount**

**Write a Pandas program to create a Pivot table and find the total sale amount region wise, manager wise, sales man wise.**

</div>

```python
class Paths:
    salesdata = '../data/SaleData.xlsx'

import pandas as pd
import numpy as np
paths = Paths()
path = paths.salesdata
df = pd.read_excel(path)
display(df)
```

<div class="alert alert-success">

**Solution 03:**
</div>

```python
display(pd.pivot_table(
    df,index=["Region","Manager","SalesMan"]
    , values="Sale_amt"
    , aggfunc='sum'
))
```
