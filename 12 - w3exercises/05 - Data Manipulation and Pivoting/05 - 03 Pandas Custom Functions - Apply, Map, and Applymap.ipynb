**16. Fill Missing Values with Custom Function Using apply()**

**Write a Pandas program to apply a custom function to fill missing values in a DataFrame.**

</div>

```python
import pandas as pd
import numpy as np

# Create date range
date_range = pd.date_range(start='2023-01-01', periods=35, freq='D')

# Create a dictionary of data
data = {
    'date': date_range,
    'customer_id': np.random.randint(1000, 2000, size=35),
    'product_id': np.random.randint(100, 200, size=35),
    'quantity': np.random.randint(1, 10, size=35),
    'price': np.round(np.random.uniform(10.0, 100.0, size=35), 2),
    'discount': np.round(np.random.uniform(0, 0.3, size=35), 2),
    'revenue': lambda df: df['quantity'] * df['price'] * (1 - df['discount']),
    'store_id': np.random.randint(1, 10, size=35),
    'region': np.random.choice(['North', 'South', 'East', 'West'], size=35),
    'payment_method': np.random.choice(['Credit Card', 'Cash', 'Online'], size=35),
    'is_member': np.random.choice([True, False], size=35),
    'delivery_days': np.random.randint(1, 7, size=35),
    'customer_age': np.random.randint(18, 70, size=35),
    'gender': np.random.choice(['Male', 'Female', 'Other'], size=35),
    'product_category': np.random.choice(['Electronics', 'Clothing', 'Books', 'Food'], size=35),
    'rating': np.round(np.random.uniform(1.0, 5.0, size=35), 1),
    'return_status': np.random.choice(['Returned', 'Not Returned'], size=35),
    'feedback': np.random.choice(['Positive', 'Neutral', 'Negative'], size=35),
}

# Create DataFrame
df = pd.DataFrame(data)

# Compute revenue using the lambda function above
df['revenue'] = df['quantity'] * df['price'] * (1 - df['discount'])

# Add 'shipping_cost' with some missing values
df['shipping_cost'] = np.where(np.random.rand(len(df)) < 0.2, np.nan, np.round(np.random.uniform(5.0, 20.0, size=len(df)), 2))

# Add 'coupon_code' with some missing values
coupon_options = ['SAVE10', 'WELCOME5', 'FREESHIP', np.nan]
df['coupon_code'] = np.random.choice(coupon_options, size=len(df), p=[0.2, 0.2, 0.2, 0.4])

# Add 'delivery_rating' with some missing values
df['delivery_rating'] = np.where(np.random.rand(len(df)) < 0.3, np.nan, np.round(np.random.uniform(1.0, 5.0, size=len(df)), 1))

# Display the updated DataFrame
display(df)
```

```python
# coupon_code has nan values, not NaN
# we will first make them NaN
df['coupon_code'] = df['coupon_code'].replace('nan', np.nan)
```

```python
# 16. Fill Missing Values with Custom Function Using apply()
# Write a Pandas program to apply a custom function to fill missing values in a DataFrame.

def fill_missing_values(row):
    """
    Fill missing values in the 'shipping_cost', 'coupon_code', and 'delivery_rating' columns.
    """
    if pd.isna(row['shipping_cost']):
        row['shipping_cost'] = 10.0  # Default shipping cost
    if pd.isna(row['coupon_code']):
        row['coupon_code'] = 'NO_COUPON'
    if pd.isna(row['delivery_rating']):
        row['delivery_rating'] = 3.0  # Default delivery rating
    return row

# Apply the custom function to fill missing values
df = df.apply(fill_missing_values, axis=1)
# Display the DataFrame after filling missing values
display(df)
```

```python
def normalize_value(x, series):
    col_min = series.min()
    col_max = series.max()
    return (x - col_min) / (col_max - col_min)

# apply the normalization function to the 'price' column
df['price_normalized'] = df['price'].apply(normalize_value, args=(df['price'],))
display(df[['price', 'price_normalized']].head(10))
```

```python
# 19. Aggregate DataFrame Groups with a Custom Function
# Write a Pandas program that applies a custom function to aggregate DataFrame groups.
def aggregate_by_region(df):
    """
    Aggregate the DataFrame by 'region' and calculate the total revenue and average price.
    """
    return df.groupby('region').agg({
        'revenue': 'sum',
        'price': 'mean'
    }).reset_index()

# Apply the aggregation function
df_aggregated = aggregate_by_region(df)
display(df_aggregated)
```

```python
def agg_groups(group):
    """
    Custom aggregation function to calculate total revenue and average price for each group.
    """
    return pd.Series({
        'total_revenue': group['revenue'].sum()
        , 'average_price': group['price'].mean()
        , 'total_quantity': group['quantity'].sum()
        , 'average_discount': group['discount'].mean()
        , 'price_plusg_shipping': group['price'].mean() + group['shipping_cost'].mean()
    })

# Apply the custom aggregation function to each group
df_aggregated_custom = df.groupby(by=['region']).apply(agg_groups, include_groups=False).reset_index()
display(df_aggregated_custom)
```

```python
import pandas as pd
import numpy as np

# Define a custom function to standardize each element (z-score), handling NaN
def standardize(x, mean, std):
    if pd.isna(x):
        return np.nan
    return (x - mean) / std if std != 0 else 0  # Avoid division by zero

# Apply standardization column-wise, using nanmean and nanstd
df_numeric = df.select_dtypes(include=np.number)
means = df_numeric.apply(np.nanmean)
stds = df_numeric.apply(np.nanstd)
df_standardized = df_numeric.apply(lambda col: col.apply(standardize, args=(means[col.name], stds[col.name])))

display(df_standardized)
```
