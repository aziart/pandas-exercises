**3. Append Rows to Existing DataFrame**

**5. Join Rows and Merge with Another DataFrame**

**Write a Pandas program to join the two given dataframes along rows and merge with another dataframe along the common column id.**

</div>

```python
import pandas as pd
student_data1 = pd.DataFrame({
        'student_id': ['S1', 'S2', 'S3', 'S4', 'S5'],
         'name': ['Danniella Fenton', 'Ryder Storey', 'Bryce Jensen', 'Ed Bernal', 'Kwame Morin'], 
        'marks': [200, 210, 190, 222, 199]})

student_data2 = pd.DataFrame({
        'student_id': ['S4', 'S5', 'S6', 'S7', 'S8'],
        'name': ['Scarlette Fisher', 'Carla Williamson', 'Dante Morse', 'Kaiser William', 'Madeeha Preston'], 
        'marks': [201, 200, 198, 219, 201]})

exam_data = pd.DataFrame({
        'student_id': ['S1', 'S2', 'S3', 'S4', 'S5', 'S7', 'S8', 'S9', 'S10', 'S11', 'S12', 'S13'],
        'exam_id': [23, 45, 12, 67, 21, 55, 33, 14, 56, 83, 88, 12]})

print("Original DataFrames:")
display(student_data1)
display(student_data2)
display(exam_data)
```

<div class="alert alert-success">

**Solution 05:**
</div>

```python
pd.DataFrame.merge?
```

```python
print("\nJoin first two said dataframes along rows:")
result_data = pd.concat(
    objs=[student_data1, student_data2],
    axis=0,
    ignore_index=True  # Reset index
)
display(result_data)

print("\nNow join the said result_data and df_exam_data along student_id:")
final_merged_data = pd.merge(
    left=result_data,
    right=exam_data,
    how='left',  # Use 'left' join to keep all records from result_data
    on='student_id'  # Join on student_id
)
display(final_merged_data)
```

**7. Join with Matching Records (Inner Join)**

**Write a Pandas program to join the two dataframes with matching records from both sides where available.**

</div>

```python
import pandas as pd
student_data1 = pd.DataFrame({
        'student_id': ['S1', 'S2', 'S3', 'S4', 'S5'],
         'name': ['Danniella Fenton', 'Ryder Storey', 'Bryce Jensen', 'Ed Bernal', 'Kwame Morin'], 
        'marks': [200, 210, 190, 222, 199]})

student_data2 = pd.DataFrame({
        'student_id': ['S4', 'S5', 'S6', 'S7', 'S8'],
        'name': ['Scarlette Fisher', 'Carla Williamson', 'Dante Morse', 'Kaiser William', 'Madeeha Preston'], 
        'marks': [201, 200, 198, 219, 201]})

print("Original DataFrames:")
display(student_data1)
display(student_data2)
```

<div class="alert alert-success">

**Solution 07:**
</div>

```python
merged_data = pd.merge(
    left=student_data1
    , right=student_data2
    , on='student_id'
    , how='outer'  # Use 'outer' join to keep all records from both
    , indicator=True  # Add indicator column to show source of each row
    , suffixes=('_left', '_right')  # Add suffixes to distinguish columns from both DataFrames
)

print("Merged data (outer join):")
display(merged_data)
```

**8. Left Join Using Keys from Left DataFrame**

**Write a Pandas program to join (left join) the two dataframes using keys from left dataframe only.**

</div>

```python
import pandas as pd
data1 = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                     'P': ['P0', 'P1', 'P2', 'P3'],
                     'Q': ['Q0', 'Q1', 'Q2', 'Q3']}) 
data2 = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                      'key2': ['K0', 'K0', 'K0', 'K0'],
                      'R': ['R0', 'R1', 'R2', 'R3'],
                      'S': ['S0', 'S1', 'S2', 'S3']})
print("Original DataFrames:")
display(data1)
print("--------------------")
display(data2)
```

<div class="alert alert-success">

**Solution 08:**
</div>

```python
print("\nMerged Data (keys from data1):")
merged_data = pd.merge(
    left=data1
    , right=data2
    , how='left'  # Use 'left' join to keep all records from data
    , on=['key1', 'key2']  # Join on both key1 and key2
    , suffixes=('_left', '_right')  # Add suffixes to distinguish columns from both DataFrames
    , indicator=True  # Add indicator column to show source of each row
)
display(merged_data)
print("\nMerged Data (keys from data2):")
merged_data = pd.merge(
    left=data2
    , right=data1
    , how='left'  # Use 'left' join to keep all records from data
    , on=['key1', 'key2']  # Join on both key1 and key2
    , suffixes=('_left', '_right')  # Add suffixes to distinguish columns from both DataFrames
    , indicator=True  # Add indicator column to show source of each row
)
display(merged_data)
```
