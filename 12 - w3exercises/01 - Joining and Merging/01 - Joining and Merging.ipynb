**1. Join DataFrames along Rows (Join Rows)**

**Write a Pandas program to join the two given dataframes along rows and assign all data.**

</div>

```python
import pandas as pd

student_data1 = pd.DataFrame({
        'student_id': ['S1', 'S2', 'S3', 'S4', 'S5'],
         'name': ['Danniella Fenton', 'Ryder Storey', 'Bryce Jensen', 'Ed Bernal', 'Kwame Morin'], 
        'marks': [200, 210, 190, 222, 199]})

student_data2 = pd.DataFrame({
        'student_id': ['S4', 'S5', 'S6', 'S7', 'S8'],
        'name': ['Scarlette Fisher', 'Carla Williamson', 'Dante Morse', 'Kaiser William', 'Madeeha Preston'], 
        'marks': [201, 200, 198, 219, 201]})

print("Original DataFrames:")
display(student_data1)
print("-------------------------------------")
display(student_data2)
```

<div class="alert alert-success">

**Solution 01:**
</div>

```python
pd.concat?
```

```python
print("\nJoin the said two dataframes along rows:")

result_data = pd.concat(
    objs=[student_data1, student_data2]
    , axis=0
    , ignore_index=True # default is False, which keeps the original index
)
display(result_data)
```

```python
import pandas as pd

student_data1 = pd.DataFrame({
        'student_id': ['S1', 'S2', 'S3', 'S4', 'S5'],
         'name': ['Danniella Fenton', 'Ryder Storey', 'Bryce Jensen', 'Ed Bernal', 'Kwame Morin'], 
        'marks': [200, 210, 190, 222, 199]})

student_data2 = pd.DataFrame({
        'student_id': ['S4', 'S5', 'S6', 'S7', 'S8'],
        'name': ['Scarlette Fisher', 'Carla Williamson', 'Dante Morse', 'Kaiser William', 'Madeeha Preston'], 
        'marks': [201, 200, 198, 219, 201]})

student_data3 = pd.DataFrame({
        'student_id': ['S4', 'S5', 'S6', 'S7', 'S8', 'added_row'],
        'name': ['Scarlette Fisher', 'Carla Williamson', 'Dante Morse', 'Kaiser William', 'Madeeha Preston', 'added_row'], 
        'marks': [201, 200, 198, 219, 201, 'added_row']})

print("Original DataFrames:")
display(student_data1)
print("-------------------------------------")
display(student_data2)
display(student_data3)

print("\nJoin dataframes with different number of rows:")
result_data = pd.concat(
    objs=[student_data1, student_data3],
    axis=1,  # Change to 1 to concatenate along columns
    ignore_index=False  # Keep the original index
)
display(result_data)
```
