**1. Merge on Single Column**

**Write a Pandas program to merge two DataFrames on a single column.**

</div>

```python
import pandas as pd

# Create two sample DataFrames
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Selena', 'Annabel', 'Caeso']
})

df2 = pd.DataFrame({
    'ID': [2, 3, 4],
    'Age': [25, 30, 22]
})

display(df1)
display(df2)
```

<div class="alert alert-success">

**Solution 01:**
</div>

```python
# Merge the two DataFrames on the 'ID' column
merged_df = pd.merge(df1, df2, on='ID')

# Output the result
display(merged_df)
```

**2. Outer Join Merge**

**Write a Pandas program to perform an outer join on two DataFrames.**

</div>

```python
import pandas as pd

# Create two sample DataFrames
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Selena', 'Annabel', 'Caeso']
})

df2 = pd.DataFrame({
    'ID': [2, 3, 4],
    'Age': [25, 30, 22]
})

display(df1)
display(df2)
```

<div class="alert alert-success">

**Solution 02:**
</div>

```python
# Perform an outer join on the 'ID' column
outer_joined_df = pd.merge(df1, df2, on='ID', how='outer')

# Output the result
display(outer_joined_df)
```

**3. Left Join Merge**

**Write a Pandas program that performs a left join of two DataFrames.**

</div>

```python
import pandas as pd

# Create two sample DataFrames
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Selena', 'Annabel', 'Caeso']
})

df2 = pd.DataFrame({
    'ID': [2, 3, 4],
    'Age': [25, 30, 22]
})

display(df1)
display(df2)
```

<div class="alert alert-success">

**Solution 03:**
</div>

```python
# Perform a left join on the 'ID' column
left_joined_df = pd.merge(df1, df2, on='ID', how='left')

# Output the result
display(left_joined_df)
```

**4. Right Join Merge**

**Write a Pandas program that performs a right join of two DataFrames.**

</div>

```python
import pandas as pd

# Create two sample DataFrames
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Selena', 'Annabel', 'Caeso']
    })

df2 = pd.DataFrame({
    'ID': [2, 3, 4],
    'Age': [25, 30, 22]
})

display(df1)
display(df2)
```

<div class="alert alert-success">

**Solution 04:**
</div>

```python
# Perform a right join on the 'ID' column
right_joined_df = pd.merge(df1, df2, on='ID', how='right')

# Output the result
display(right_joined_df)
```
