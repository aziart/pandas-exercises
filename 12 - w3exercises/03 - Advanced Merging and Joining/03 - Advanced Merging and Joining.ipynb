**1. Merge on Single Column**

**Write a Pandas program to merge two DataFrames on a single column.**

</div>

```python
import pandas as pd

# Create two sample DataFrames
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Selena', 'Annabel', 'Caeso']
})

df2 = pd.DataFrame({
    'ID': [2, 3, 4],
    'Age': [25, 30, 22]
})

display(df1)
display(df2)
```

<div class="alert alert-success">

**Solution 01:**
</div>

```python
# Merge the two DataFrames on the 'ID' column
merged_df = pd.merge(df1, df2, on='ID')

# Output the result
display(merged_df)
```

**2. Outer Join Merge**

**Write a Pandas program to perform an outer join on two DataFrames.**

</div>

```python
import pandas as pd

# Create two sample DataFrames
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Selena', 'Annabel', 'Caeso']
})

df2 = pd.DataFrame({
    'ID': [2, 3, 4],
    'Age': [25, 30, 22]
})

display(df1)
display(df2)
```

<div class="alert alert-success">

**Solution 02:**
</div>

```python
# Perform an outer join on the 'ID' column
outer_joined_df = pd.merge(df1, df2, on='ID', how='outer')

# Output the result
display(outer_joined_df)
```

**5. Merge on Multiple Columns**

**Write a Pandas program to merge two DataFrames on multiple columns.**

</div>

```python
import pandas as pd

# Create two sample DataFrames
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Selena', 'Annabel', 'Caeso'],
    'Age': [25, 30, 22]
})

df2 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Selena', 'Annabel', 'Caeso'],
    'Salary': [50000, 60000, 70000]
})
display(df1)
display(df2)
```

<div class="alert alert-success">

**Solution 05:**
</div>

```python
# Merge the DataFrames on both 'ID' and 'Name' columns
merged_df = pd.merge(df1, df2, on=['ID', 'Name'])

# Output the result
display(merged_df)
```

**6. Merge Overlapping Columns**

**Write a Pandas program that merges DataFrames with overlapping column names.**

</div>

```python
# Create two sample DataFrames with overlapping column names
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Selena', 'Annabel', 'Caeso'],
    'Value': [100, 200, 300]
})

df2 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Selena', 'Annabel', 'Caeso'],
    'Value': [400, 500, 600]
})
```

<div class="alert alert-success">

**Solution 06:**
</div>

```python
# Merge the DataFrames, adding suffixes to overlapping columns
merged_df = pd.merge(
    df1
    , df2
    , on=['ID', 'Name']
    , suffixes=('_left', '_right')
)

# Output the result
display(merged_df)
```

**7. Merge on Index**

**Write a Pandas program to merge two DataFrames on their indexes.**

</div>

```python
import pandas as pd

# Create two sample DataFrames with indexes
df1 = pd.DataFrame({
    'Name': ['Selena', 'Annabel', 'Caeso'],
    'Age': [25, 30, 22]
}, index=[1, 2, 3])

df2 = pd.DataFrame({
    'Salary': [50000, 60000, 70000]
}, index=[1, 2, 3])

display(df1)
display(df2)
```

<div class="alert alert-success">

**Solution 07:**
</div>

```python
# Merge the DataFrames on their indexes
merged_df = pd.merge(
    df1
    , df2
    , left_index=True
    , right_index=True
)

# Output the result
display(merged_df)
```

**8. Merge Diff Column Names**

**Write a Pandas program to merge different column names in DataFrames.**

</div>

```python
import pandas as pd

# Create two sample DataFrames with different join column names
df1 = pd.DataFrame({
    'Employee_ID': [1, 2, 3],
    'Name': ['Selena', 'Annabel', 'Caeso']
})

df2 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Salary': [50000, 60000, 70000]
})

display(df1)
display(df2)
```

<div class="alert alert-success">

**Solution 08:**
</div>

```python
# Merge the DataFrames, specifying different column names for the join
merged_df = pd.merge(df1, df2, left_on='Employee_ID', right_on='ID')

# Output the result
display(merged_df)
```
