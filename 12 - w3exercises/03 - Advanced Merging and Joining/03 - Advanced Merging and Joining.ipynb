**1. Merge on Single Column**

**Write a Pandas program to merge two DataFrames on a single column.**

</div>

```python
import pandas as pd

# Create two sample DataFrames
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Selena', 'Annabel', 'Caeso']
})

df2 = pd.DataFrame({
    'ID': [2, 3, 4],
    'Age': [25, 30, 22]
})

display(df1)
display(df2)
```

<div class="alert alert-success">

**Solution 01:**
</div>

```python
# Merge the two DataFrames on the 'ID' column
merged_df = pd.merge(df1, df2, on='ID')

# Output the result
display(merged_df)
```

**9. Merge Duplicate Keys**

**Write a Pandas program to merge DataFrames with duplicate Keys.**

</div>

```python
import pandas as pd

# Create two sample DataFrames with duplicate keys
df1 = pd.DataFrame({
    'ID': [1, 1, 2],
    'Name': ['Annabel', 'Annabel', 'Selena']
})

df2 = pd.DataFrame({
    'ID': [1, 2],
    'Age': [25, 30],
    'Name': ['Annabel', 'Selena']
})

display(df1)
display(df2)
```

<div class="alert alert-success">

**Solution 09:**
</div>

```python
# Merge the DataFrames with duplicate keys
merged_df = pd.merge(
    left=df1
    , right=df2
    # , on='ID'
    , left_on='ID'
    , right_on='ID'
    , how='left'
    , suffixes=('_df1', '_df2')
).drop_duplicates()

# Output the result
display(merged_df)
```

**10. Merge Multiple DataFrames**

**Write a Pandas program to merge multiple DataFrames on a common column.**

</div>

```python
import pandas as pd

# Create three sample DataFrames
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Selena', 'Annabel', 'Caeso']
})

df2 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Age': [25, 30, 22]
})

df3 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Salary': [50000, 60000, 70000]
})

display(df1)
display(df2)
display(df3)
```

<div class="alert alert-success">

**Solution 10:**
</div>

```python
# Merge the three DataFrames on the 'ID' column
merged_df = pd.merge(
    left = pd.merge(df1, df2, on='ID')
    , right = df3
    , on='ID'
)

# Output the result
display(merged_df)
```

**15. Merge with Custom Sorting**

**Write a Pandas program to merge DataFrames with custom sorting.**

</div>

```python
import pandas as pd
# Create two sample DataFrames
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Selena', 'Annabel', 'Caeso']
})

df2 = pd.DataFrame({
    'ID': [2, 3, 1],
    'Age': [30, 22, 25]
})

display(df1)
display(df2)
```

<div class="alert alert-success">

**Solution 15:**
</div>

```python
# Merge the DataFrames on the 'ID' column
merged_df = pd.merge(
    left=df1
    , right=df2
    , on='ID'
)

# Sort the result by 'Age' column
sorted_df = merged_df.sort_values(by='Age')

# Output the result
display(sorted_df)
```
