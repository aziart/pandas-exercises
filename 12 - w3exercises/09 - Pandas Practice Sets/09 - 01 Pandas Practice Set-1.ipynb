**26. Calculate Mean of Each Numeric Column**

**Write a Pandas program to calculate the mean of each numeric column of diamonds DataFrame.**

</div>

```python
path = '../data/'
filename = 'diamonds.csv'
df = pd.read_csv(path + filename)
display(df.head(20))
```

<div class="alert alert-success">

**Solution 26:**
</div>

```python
# 26. Calculate Mean of Each Numeric Column
# Write a Pandas program to calculate the mean of each numeric column of diamonds DataFrame.
# simple

df.select_dtypes(include=np.number).describe().style.format("{:,.1f}")
```

```python
# 26. Calculate Mean of Each Numeric Column
# Write a Pandas program to calculate the mean of each numeric column of diamonds DataFrame.

def metrics(group, col):
    return pd.Series({
        'mean': group[col].mean()
        , 'count': group[col].count()
    })


group_col = "cut"  # the column we group by

for col in df.select_dtypes(include=np.number).columns:
    if col == group_col:  # skip grouping column itself
        continue
    df_summary = df.groupby(group_col).apply(lambda g: metrics(g, col), include_groups=False).reset_index()
    print("Mean and Count of {} feature".format(col))
    df_summary.insert(1, "feature", col)  # add column name for clarity
    display(df_summary)
```

**28. Calculate Mean of Price for Each Cut**

**Write a Pandas program to calculate the mean of price for each cut of diamonds DataFrame.**

</div>

```python
path = '../data/'
filename = 'diamonds.csv'
df = pd.read_csv(path + filename)
display(df.head(20))
```

<div class="alert alert-success">

**Solution 28:**
</div>

```python
# 28. Calculate Mean of Price for Each Cut
# Write a Pandas program to calculate the mean of price for each cut of diamonds DataFrame.

group_col = "cut"  # the column we group by
column_used = "price"

def metrics(group):
    return pd.Series({
        'mean': group[column_used].mean()
        , 'min': group[column_used].min()
        , 'max': group[column_used].max()
    })


df_summary = df.groupby(by=[group_col]).apply(metrics, include_groups=False)

df_summary.insert(0, "feature", column_used)  # add column name for clarity
display(df_summary)


# alternative

df_summary = df.groupby(by=['cut']).agg(
    avg = ('price', 'mean')
)
df_summary.style.format("{:,.2f}")
```

**29. Calculate Count, Minimum, and Maximum Price for Each Cut**

**Write a Pandas program to calculate count, minimum, maximum price for each cut of diamonds DataFrame.**

</div>

```python
path = '../data/'
filename = 'diamonds.csv'
df = pd.read_csv(path + filename)
display(df.head(20))
```

<div class="alert alert-success">

**Solution 29:**
</div>

```python
# 29. Calculate Count, Minimum, and Maximum Price for Each Cut
# Write a Pandas program to calculate count, minimum, maximum price for each cut of diamonds DataFrame.


group_col = "cut"  # the column we group by
column_used = "price"

def metrics(group):
    return pd.Series({
        'count': group[column_used].count()
        , 'mean': group[column_used].mean()
        , 'min': group[column_used].min()
        , 'max': group[column_used].max()
    })


df_summary = df.groupby(by=[group_col]).apply(metrics, include_groups=False)

df_summary.insert(0, "feature", column_used)  # add column name for clarity
display(df_summary)
```

```python
# 30. Create Side-by-Side Bar plot
# Write a Pandas program to create a side-by-side bar plot of the diamonds DataFrame.

group_col = "cut"  # the column we group by
column_used = "price"
df_summary = df.groupby(by=[group_col]).agg(
    avg = (column_used, 'mean')
    , min = (column_used, 'min')
    , max = (column_used, 'max')
)
df_summary.plot.bar(rot=0)
plt.title("Average, Minimum, and Maximum Price for Each Cut")
plt.xlabel("Cut")
plt.ylabel("Price (USD)")
plt.show()
```

**Write a Pandas program to count how many times each value in cut series of diamonds DataFrame occurs.**

<div class="alert alert-success">

**Solution 31:**

```python
# 31. Count Occurrences in 'cut' Series
# Write a Pandas program to count how many times each value in cut series of diamonds DataFrame occurs.

display(df.groupby(by=['cut']).size())
df.groupby(by=['cut']).size().plot.bar(rot=45)
```

```python
df_summary = df.groupby(by=['cut']).agg(
    cnt = ('cut','count')
).reset_index().melt(id_vars=['cut'], value_vars=['cnt'], var_name='feature', value_name='value')

display(df_summary)

g = sns.barplot(
    data=df_summary,
    x="cut",
    y="value",
    hue="feature",
    legend=True,
    alpha=0.5
);
```

```python
df_summary = df.groupby(by=['cut']).agg(
    cnt = ('cut','count')
    , avg = ('price', 'mean')
    , min = ('price', 'min')
    , max = ('price', 'max')
).reset_index().melt(id_vars=['cut'], value_vars=['cnt', 'min', 'avg', 'max'], var_name='feature', value_name='value')

display(df_summary)

g = sns.barplot(
    data=df_summary,
    x="cut",
    y="value",
    hue="feature",
    legend=True,
    alpha=0.5
);
```
