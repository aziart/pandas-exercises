**Exercise 8:**

**Group a DataFrame by a column and calculate the mean of each group.**

</div>

```python
path = '../data/'
file_name = 'titanic.csv'
df = pd.read_csv(path + file_name)
df.head(20)
```

<div class="alert alert-success">

**Solution 08:**
</div>

```python
# Exercise 8:
# Group a DataFrame by a column and calculate the mean of each group.

# simple way
df.groupby(by=['who'])[['survived']].mean()
```

```python
# more complicated one
def calculate_several_metrics(group):
    return pd.Series(
        {
            'mean_survived': group['survived'].mean()
            , 'max_age': group['age'].max()
            , 'min_fare': group['fare'].min()
            , 'std_age': group['age'].std()
            , 'average_age_of_survivors': group.loc[group['survived'] == 1, 'age'].mean()
        }
    )

df.groupby(by=['who']).apply(
    calculate_several_metrics
    , include_groups=False
)
```

**Exercise 23:**

**Filter rows by multiple conditions.**

</div>

```python
path = '../data/'
file_name = 'titanic.csv'
df = pd.read_csv(path + file_name)
print(df.shape)
df.head(20)
```

<div class="alert alert-success">

**Solution 23:**
</div>

```python
# Exercise 23:
# Filter rows by multiple conditions.


f1 = (df['sex']=='female')
f2 = (df['fare']>=10)

# filter + groupby + apply custom functions

def calculate_values(group):
    return pd.Series(
        {
            'age_sum': group['age'].sum()
            , 'alone_true': group.loc[group['alone']==True, 'alone'].sum()
            , 'alone_perc': group.loc[group['alone']==True, 'alone'].count() / group['alone'].count()
        }
    )

df[f1 & f2].groupby(by=['pclass']).apply(
    calculate_values
    , include_groups=False
)
```

```python
# Exercise 90:
# Create a DataFrame with datetime index and calculate the expanding mean for each group.

date_rng = pd.date_range(start='2023-01-01', end='2023-12-31', freq='D')
df = pd.DataFrame(date_rng, columns=['date'])
df['data'] = np.random.rand(len(date_rng))
df['group'] = np.random.choice(['A', 'B', 'C'], size=len(date_rng))
df.set_index('date', inplace=True)
print("DataFrame with datetime index:")
display(df.head(10))
print("Expanding mean for each group:")
# expanding_mean_grouped = df.groupby(by=['group'])['data'].expanding().mean().reset_index(drop=True)
# df['expanding_mean_grouped'] = expanding_mean_grouped
# display(df.head(20))

# fix using transform
expanding_mean_grouped = df.groupby('group')['data'].transform(lambda x: x.expanding().mean())
df['expanding_mean_grouped'] = expanding_mean_grouped
display(df.head(20))
```
