**Exercise 5:**

**Add a new column to an existing DataFrame.**

</div>

```python
path = '../data/'
file_name = 'titanic.csv'
df = pd.read_csv(path + file_name)
df.head(20)
```

<div class="alert alert-success">

**Solution 05:**
</div>

```python
# Exercise 5:
# Add a new column to an existing DataFrame.

def adding_new_column(row):
    if row['who'] == 'man':
        return 1
    elif row['who'] == 'woman':
        return 2
    else:
        return 0

df['encoded_gender'] = df.apply(adding_new_column, axis=1)

#make sure all genders encoded
df.groupby(by=['who', 'encoded_gender']).agg(
    count = ('who','count')
    , sum_fare = ('fare','sum')
)
```

</div>

```python
# choosing column to drop
column2drop = df.isnull().sum()[df.isnull().sum().values == df.isnull().sum().max()].index.values[0]

# dropping the column
df.drop(
    columns=column2drop
    , axis=1
    , inplace=True
)

df.head(20)
```

**Exercise 8:**

**Group a DataFrame by a column and calculate the mean of each group.**

</div>

```python
path = '../data/'
file_name = 'titanic.csv'
df = pd.read_csv(path + file_name)
df.head(20)
```

<div class="alert alert-success">

**Solution 08:**
</div>

```python
# Exercise 8:
# Group a DataFrame by a column and calculate the mean of each group.

# simple way
df.groupby(by=['who'])[['survived']].mean()
```

```python
# more complicated one
def calculate_several_metrics(group):
    return pd.Series(
        {
            'mean_survived': group['survived'].mean()
            , 'max_age': group['age'].max()
            , 'min_fare': group['fare'].min()
            , 'std_age': group['age'].std()
            , 'average_age_of_survivors': group.loc[group['survived'] == 1, 'age'].mean()
        }
    )

df.groupby(by=['who']).apply(
    calculate_several_metrics
    , include_groups=False
)
```
