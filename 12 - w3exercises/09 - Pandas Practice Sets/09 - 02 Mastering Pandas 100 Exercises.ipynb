```python
# Exercise 5:
# Add a new column to an existing DataFrame.

def adding_new_column(row):
    if row['who'] == 'man':
        return 1
    elif row['who'] == 'woman':
        return 2
    else:
        return 0

df['encoded_gender'] = df.apply(adding_new_column, axis=1)

#make sure all genders encoded
df.groupby(by=['who', 'encoded_gender']).agg(
    count = ('who','count')
    , sum_fare = ('fare','sum')
)
```

**Exercise 8:**

**Group a DataFrame by a column and calculate the mean of each group.**

</div>

```python
path = '../data/'
file_name = 'titanic.csv'
df = pd.read_csv(path + file_name)
df.head(20)
```

<div class="alert alert-success">

**Solution 08:**
</div>

```python
# Exercise 8:
# Group a DataFrame by a column and calculate the mean of each group.

# simple way
df.groupby(by=['who'])[['survived']].mean()
```

```python
# more complicated one
def calculate_several_metrics(group):
    return pd.Series(
        {
            'mean_survived': group['survived'].mean()
            , 'max_age': group['age'].max()
            , 'min_fare': group['fare'].min()
            , 'std_age': group['age'].std()
            , 'average_age_of_survivors': group.loc[group['survived'] == 1, 'age'].mean()
        }
    )

df.groupby(by=['who']).apply(
    calculate_several_metrics
    , include_groups=False
)
```

```python
# Exercise 22:
# Rename columns in a DataFrame.


print("DataFrame before renaming columns:")
display(df.head(20))

rename_names = {}
for i in df.columns:
    rename_names[i] = i.upper()
print('Rename columns dictionary:')
display(rename_names)


df.rename(
    columns=rename_names
    , inplace=True
)
print("DataFrame after renamin columns:")
display(df.head(20))
```

**Exercise 23:**

**Filter rows by multiple conditions.**

</div>

```python
path = '../data/'
file_name = 'titanic.csv'
df = pd.read_csv(path + file_name)
print(df.shape)
df.head(20)
```

<div class="alert alert-success">

**Solution 23:**
</div>

```python
# Exercise 23:
# Filter rows by multiple conditions.


f1 = (df['sex']=='female')
f2 = (df['fare']>=10)

# filter + groupby + apply custom functions

def calculate_values(group):
    return pd.Series(
        {
            'age_sum': group['age'].sum()
            , 'alone_true': group.loc[group['alone']==True, 'alone'].sum()
            , 'alone_perc': group.loc[group['alone']==True, 'alone'].count() / group['alone'].count()
        }
    )

df[f1 & f2].groupby(by=['pclass']).apply(
    calculate_values
    , include_groups=False
)
```
