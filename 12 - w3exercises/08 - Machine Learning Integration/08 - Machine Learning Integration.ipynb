**4. Filling Missing Values with the Mean**

**Write a Pandas program that fills missing values with the Mean.**

</div>

```python
path = '../data/'
filename = 'titanic.csv'

df = pd.read_csv(path + filename)
# add some missing values
df.loc[5:9, 'age'] = np.nan
df.drop(columns=['Unnamed: 15'], inplace=True)
df.head(20)
```

<div class="alert alert-success">

**Solution 04:**
</div>

```python
# 4. Filling Missing Values with the Mean
# Write a Pandas program that fills missing values with the Mean.

df_filled_mean = df.copy()
df_filled_mean['age'].fillna(
    df_filled_mean['age'].mean()
    , inplace=True
)

df_filled_mean.head(20)
```

```python
df = pd.read_csv(path + filename)
# add some missing values
df.loc[5:9, 'age'] = np.nan
df.drop(columns=['Unnamed: 15'], inplace=True)
df.head(20)


df_filled_mean = df.copy()
df_filled_mean.fillna(
    {'age': df_filled_mean['age'].mean()}
    , inplace=True
)
df_filled_mean.head(20)
```

```python
df = pd.read_csv(path + filename)
# add some missing values
df.loc[5:9, 'age'] = np.nan
df.drop(columns=['Unnamed: 15'], inplace=True)
df.head(20)


df_filled_mean = df.copy()
df_filled_mean['age'] = df_filled_mean['age'].fillna(df_filled_mean['age'].mean())
df_filled_mean.head(20)
```
