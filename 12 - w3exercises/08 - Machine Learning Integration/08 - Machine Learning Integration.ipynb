**Write a Pandas program to drop rows with missing values from a dataset.**

```python
path = '../data/'
filename = 'titanic.csv'

df = pd.read_csv(path + filename)
# add some missing values
df.loc[5:9, 'age'] = np.nan
df.drop(columns=['Unnamed: 15'], inplace=True)
df.head(20)
```

<div class="alert alert-success">

**Solution 03:**
</div>

```python
# 3. Dropping Rows with Missing Values from a Dataset
# Write a Pandas program to drop rows with missing values from a dataset.


df_non_missing = df.dropna().copy()
df_non_missing.head(20)
```

```python
df_non_missing_age = df.dropna(subset=['age']).copy()
df_non_missing_age.head(20)
```
