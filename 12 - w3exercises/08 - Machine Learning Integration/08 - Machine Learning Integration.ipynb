**10. Removing Outliers from a Dataset**

**Write a Pandas program that removes outliers from a Dataset.**

</div>

```python
path = '../data/'
filename = 'titanic.csv'

df = pd.read_csv(path + filename)
# add some missing values
df.loc[5:9, 'age'] = np.nan
df.drop(columns=['Unnamed: 15'], inplace=True)
df.head(20)
```

<div class="alert alert-success">

**Solution 10:**
</div>

```python
# 10. Removing Outliers from a Dataset
# Write a Pandas program that removes outliers from a Dataset.


df_no_outliers = df.copy()
print("Before removing outliers:", df_no_outliers.shape)

def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Apply outlier removal for all numeric columns
for column in df_no_outliers.select_dtypes(include=['float64', 'int64']).columns:
    df_no_outliers = remove_outliers(df_no_outliers, column)

df_no_outliers.reset_index(drop=True, inplace=True)
print("After removing outliers:", df_no_outliers.shape)
display(df_no_outliers.head(20))
```

```python
# Remove outliers from the 'age' column using Z-scores
from scipy import stats

# Drop NaN before computing z-scores
ages = df['age'].dropna()

# Compute Z-scores
z_scores = np.abs(stats.zscore(ages))

# Filter out outliers
df_cleaned = df.loc[ages.index[z_scores < 3]]

print("Before removing outliers:", df.shape)
print("After removing outliers:", df_cleaned.shape)
display(df_cleaned.head(20))
```

```python
# 11. Imputing Missing Values Using K-Nearest Neighbours
# Write a Pandas program that imputes missing values using K-Nearest neighbours.

from sklearn.impute import KNNImputer

# Separate the numeric columns (Age and Salary) from non-numeric ones (Name, Gender)
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
non_numeric_cols = df.select_dtypes(include=['object']).columns

# Apply KNN imputation only to the numeric columns
imputer = KNNImputer(n_neighbors=3)
df_numeric_imputed = pd.DataFrame(imputer.fit_transform(df[numeric_cols]), columns=numeric_cols)

# Combine the non-numeric columns with the imputed numeric data
df_imputed = pd.concat([df[non_numeric_cols].reset_index(drop=True), df_numeric_imputed], axis=1)

# Output the dataset with imputed values
display(df_imputed.head(20))
```

**12. Selecting Features Using Variance Threshold**

**Write a Pandas program to select feature selection using variance threshold.**

</div>

```python
path = '../data/'
filename = 'titanic.csv'

df = pd.read_csv(path + filename)
# add some missing values
df.loc[5:9, 'age'] = np.nan
df.drop(columns=['Unnamed: 15'], inplace=True)
df.head(20)
```

<div class="alert alert-success">

```python
# 12. Selecting Features Using Variance Threshold
# Write a Pandas program to select feature selection using variance threshold.



from sklearn.feature_selection import VarianceThreshold

# Select only numeric columns
df_numeric = df.select_dtypes(include=['number'])

# Create a VarianceThreshold instance
selector = VarianceThreshold(threshold=0.1)

print("Before feature selection:", df_numeric.shape)

# Fit and transform the numeric data
df_selected = selector.fit_transform(df_numeric)

# Convert back to DataFrame with selected features
df_selected = pd.DataFrame(
    df_selected, 
    columns=df_numeric.columns[selector.get_support()],
    index=df.index
)

print("After feature selection:", df_selected.shape)

# Display the selected features
display(df_selected.head(20))
```

```python
# 13. Handling Class Imbalance Using Random Oversampling
# Write a Pandas program to handling class imbalance using random oversampling.

from imblearn.over_sampling import RandomOverSampler
# Separate features and target
X = df.drop(columns=['survived'])
y = df['survived']

# Apply random oversampling
ros = RandomOverSampler(random_state=42)
X_resampled, y_resampled = ros.fit_resample(X, y)
print("Before oversampling:", y.value_counts().to_dict())
print("After oversampling:", pd.Series(y_resampled).value_counts().to_dict())
```
