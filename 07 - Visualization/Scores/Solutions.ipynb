### Step 1. Import the necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="white")
sns.set_context("paper")
import warnings
warnings.filterwarnings("ignore")

pd.set_option("display.float_format", "{:.2f}".format)
pd.set_option("display.max_columns", None)

%matplotlib inline
```

### Step 2. Create the DataFrame it should look like below.

```python
data = {'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'], 
            'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'], 
            'female': [0, 1, 1, 0, 1],
            'age': [42, 52, 36, 24, 73], 
            'preTestScore': [4, 24, 31, 2, 3],
            'postTestScore': [25, 94, 57, 62, 70]}

df = pd.DataFrame(data, columns = ['first_name', 'last_name', 'age', 'female', 'preTestScore', 'postTestScore'])

df
```

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

```python
g = sns.relplot(
    data=df,
    x="preTestScore",
    y="postTestScore",
    # hue="age",
    kind="scatter",  # OR scatter
    aspect=2,
    size=np.array(df.age)
)

g.set_xticklabels(rotation=45, horizontalalignment="right", step=2);
```

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

```python
g = sns.relplot(
    data=df,
    x="preTestScore",
    y="postTestScore",
    hue="female",
    kind="scatter",  # OR scatter
    aspect=2,
    size="age",
    sizes=(100, 450),
    palette="muted"
)

g.set_xticklabels(rotation=45, horizontalalignment="right", step=2);
```
