### Step 1. Import the necessary libraries:

```python
import numpy as np
import pandas as pd
pd.set_option("display.float_format", "{:.2f}".format)
pd.set_option("display.max_columns", None)

import warnings
warnings.filterwarnings("ignore")

# visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns


# print the graphs in the notebook
%matplotlib inline

# set seaborn style to white
sns.set_style("white")
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv). 

```python
path = "../../data/"
fname = "tips.csv"
df = pd.read_csv(path+fname)
df
```

### Step 3. Assign it to a variable called tips

```python
tips = df.copy()
tips
```

### Step 4. Delete the Unnamed 0 column

```python
tips = tips.drop(columns=['Unnamed: 0']
             , axis=1)
tips
```

### Step 5. Plot the total_bill column histogram

```python
total_bill_start = -10
total_bill_end = 65
total_bill_interval = 10

buckets = np.arange(total_bill_start
                  , total_bill_end
                  , total_bill_interval)
buckets
total_bill_group = pd.cut(tips['total_bill'], buckets)
tips.insert(loc = len(tips.columns)
            , column="total_bill_group"
            , value=total_bill_group
            , allow_duplicates=True)
tips
# online_rt.groupby(UnitPrice_grouped)['SalesVolume'].sum()
```

```python
sns.histplot(
    data=tips
    , x="total_bill"
    , bins=buckets
    , kde=True
);
plt.show();

sns.histplot(
    data=tips
    , x="total_bill"
    , kde=True
);
```

```python
sns.displot(
    data=tips
    , x='total_bill'
    , kind='hist'
)
sns.displot(
    data=tips
    , x='total_bill'
    , kind='kde'
)
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
tips
```

```python
# Scatter plot + distribution

sns.jointplot(
    data=tips
    , x='total_bill'
    , y='tip'
    , hue="time"
    # , height=10
    , legend=True
    , ratio=4
    , kind="scatter"
);
```

```python
sns.jointplot(
    data=tips
    , x='total_bill'
    , y='tip'
    , kind='scatter'
    , legend=False
)
```
